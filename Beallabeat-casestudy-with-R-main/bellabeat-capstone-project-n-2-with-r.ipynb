#### Import the datasets.

# <p id = "b3"><strong><h2>PROCESS</h2></strong></p>
Time to clean and organize the different dataset in order to have a better understanding and easier access to the crucial parts of the data.


#### Cleaning the datasets.

Renaming all the names of the variables of the used datasets to lowercase for better integrity.

```r
daily_activity <- rename_with(daily_activity, tolower)
heartrate_seconds <- rename_with(heartrate_seconds, tolower)
sleep_day <- rename_with(sleep_day, tolower)
```

Cleaning the "daily_activities" datasets from eventual na values and modify the "activitydate" column to a date format.

```r
daily_activity <- daily_activity %>% 
  drop_na() %>% 
  mutate(activitydate = as.Date(activitydate, format = "%m/%d/%Y"))
```

Cleaning the "heartrate_seconds" datasets from eventual na values, modify the "Time" column to a date format and summarize the dataset by calculating the mean of the heart rate throughout the day.


```r
concise_heartrate <- heartrate_seconds %>%
  drop_na() %>% 
  separate(time , c("day","time","AM/PM")," ") %>% 
  mutate(day = as.Date(day, format = "%m/%d/%Y"))%>% 
  group_by(id, day) %>% 
  summarize(avg_heartrate = mean(value))
```

Dividing the "sleep_day" dataset into the ones that slept between the 7-9 hours of recommended sleep, and the ones who didn't, cleaning both datasets from eventual na values and modify the "sleepday" column to a date format.

```r
good_sleep <- filter(sleep_day, totalminutesasleep >= 420 & totalminutesasleep <= 540) %>% 
  drop_na() %>% 
  separate(sleepday , c("day","time","AM/PM")," ") %>% 
  mutate(day = as.Date(day, format = "%m/%d/%Y"))

bad_sleep <- filter(sleep_day,totalminutesasleep <= 420 | totalminutesasleep >= 540) %>% 
  drop_na() %>% 
  separate(sleepday , c("day","time","AM/PM")," ") %>% 
  mutate(day = as.Date(day, format = "%m/%d/%Y"))
```

<br/><br/>
#### Merging the datasets.
Merging the datasets in order to have more relevant information for the "Leaf" product and marketing analysis.

```r
heartSum_goodsleep <- merge(concise_heartrate, good_sleep)
heartSum_badsleep <- merge(concise_heartrate, bad_sleep)
activity_heartrate <-  merge(rename(daily_activity, day = activitydate), concise_heartrate)
```

First i wanted to see the correlation between steps taken and calories burned daily.

```r
ggplot(data = daily_activity) + 
  geom_point(mapping = aes(x = totalsteps, y = calories, color = calories)) +
  scale_color_gradient(low = "red", high = "yellow")+
  geom_smooth(mapping = aes(x =totalsteps, y = calories), method = "loess", formula = "y ~ x")+
  theme_light()+
  ggtitle("Total Steps vs. Calories Burned") +
  xlab("Total Steps")+
  ylab("Calories")
options(repr.plot.width = 14, repr.plot.height = 8)
```

The correlation is not the highest but it is present:

Now that I figured out what the best activity of the day should be I wanted to focus on the remaining part of the day, when people rest, and find out how they should do it.


So, next i wanted to see if bad sleeping habit would create heart complicaton for the subjects and if the "Leaf" product should be doing something about it.

```r
w <- (nrow(na.omit(bad_sleep)))
v <- (nrow(na.omit(good_sleep)))


pie <- data.frame(category = c("Good Sleep Occurrences", "Bad sleep Occurrences"),
                  count = c(w, v))
pie$fraction <-  pie$count / sum(pie$count)
pie$ymax <-  cumsum(pie$fraction)
pie$ymin <-  c(0, head(pie$ymax, n=-1))
pie$labelPosition <- (pie$ymax + pie$ymin) / 2
pie$label <- paste0(pie$category, "\n value: ", pie$count,"\n","(", round(pie$fraction*100, digits = 0),"%)")




#creo il grafico
ggplot(pie, aes(ymax=ymax, ymin=ymin, xmax=4, xmin=3, fill=category)) +
     geom_rect() +
     coord_polar(theta="y") + 
     xlim(c(2, 4))+
     geom_label( x=3.5, aes(y=labelPosition, label=label), size=4) +
     scale_fill_brewer(palette=1) +
     theme_void()+
     theme(legend.position = "none")
options(repr.plot.width = 14, repr.plot.height = 8)

```

It is obvious that the subjects don't have a particularly strict sleeping schedule with only about 46% of the times they sleep being the raccomanded amount, about 53% of their sleeping schedule is not consider healthy for an adult, and of this 53%, 2.2% is incredibily unhealthy, sleeping only for less than 90 min, and another 1% is sleeping for huge amounts of time, more than 12 hours.

```r
H_g <- ggplot(data = heartSum_goodsleep) + 
  geom_point(mapping = aes(x =totalminutesasleep , y =avg_heartrate, color = avg_heartrate ))+ 
  scale_color_gradient(low = "red", high = "yellow")+
  geom_smooth(mapping = aes(x =totalminutesasleep , y =avg_heartrate ), method = "loess", formula = y ~ x, color = "blue") + 
  theme_light() +
  ylim(50, 100)+
  ggtitle("Healthy amount of sleep: 420 - 520 Min.") +
     xlab("Total Minutes Asleep")+
     ylab("Avg Heartrate (bpm)")

H_b <- ggplot(data = heartSum_badsleep) + 
  geom_point(mapping = aes(x =totalminutesasleep , y =avg_heartrate , color = avg_heartrate)) +
  scale_color_gradient(low = "red", high = "yellow")+
  geom_smooth(mapping = aes(x =totalminutesasleep , y =avg_heartrate ), method = "loess", formula = y ~ x, color = "blue") + 
  theme_light() +
  ylim(50, 100) +
  ggtitle("Unhealthy amount of sleep: More / Less than 420 - 520 Min.") +
     xlab("Total Minutes Asleep")+
     ylab("Avg Heartrate (bpm)")


plot_grid(H_g,H_b, ncol = 1)
options(repr.plot.width = 14, repr.plot.height = 8)
```


It doesn't appear that any particularly strong correlation exists between an individual who sleeps with healthy habits and their average heart rate, meaning that if the subjects sleep regularly the heartrate will stay the same.

The majority of the subject have an healthy heart rate falling between the 60 and 100 BPM.

It is important to notice though, that while it doesn't appear to be any correlation between "Good sleep" and an abnormal average heart rate,
there is a slight inverse correlation between "Bad sleep" and average heart rate meaning that the more a subject sleeps the more is likely that its heart rate will be lower than normal.

Also both datasets present 2 outliers with an unusually high average heart rate, which get slightly higher when those subjects sleep very little.

Here the correlation coefficient of both "Good sleep" and "Bad sleep" respectably:

So I made a bar chart comparing the various days of the week in terms of sedentary minutes and very active minutes.

I choosed to compare sedentary minutes with very active minutes since we found out that very active minutes is the variable more directly correlated to calories loss and so it's the one that probably takes more effort and dedication to be consistent with.

```r
  #transformo le date in giorni della settimana
daily_activity$weekday <-  weekdays(daily_activity$activitydate)
  #ritaglio i le variabili che mi interessano
forStand_daily_activity <-  daily_activity[c("sedentaryminutes", "veryactiveminutes", "weekday")]
  
  #calcolo la deviazione standard per ogni row
standardized_daily <- forStand_daily_activity %>% mutate_at(c("sedentaryminutes", "veryactiveminutes"), ~(scale(.) %>% as.vector)) %>% group_by(weekday)



#making 2 datasets with the standardized mean of the weekday of the corrispecting variable
weekday_sed <- standardized_daily[c("sedentaryminutes", "weekday")] %>%
	group_by(weekday) %>%
	summarise(sedentarySD = mean(sedentaryminutes))

weekday_ver <- standardized_daily[c("veryactiveminutes", "weekday")] %>%
	group_by(weekday) %>%
	summarise(veryactiveSD = mean(veryactiveminutes))

#unisco i dataset
c <- merge.data.frame(weekday_sed, weekday_ver)
  #riordino i grioni della settimna
c$weekday <- factor(c$weekday, levels= c("Monday", "Tuesday", 
    "Wednesday", "Thursday", "Friday", "Satuday", "Sunday"))
c <- c[order(c$weekday), ]
  #changing the data shape so there are 2 groups "sedentaryst" and veryactivest" 
c1 <- reshape2::melt(c, id.vars = "weekday")



#making the bar plot
ggplot(c1, aes(weekday, value, fill = variable)) +
  geom_col(position = "dodge")+
  scale_fill_brewer(palette=1)+
  theme_minimal()
options(repr.plot.width = 14, repr.plot.height = 8)
```

With the bar chart, which was standardized since sedentary minutes and very active minutes have much differents average values, we can clearly see that there are some days worst than others in terms of motivation.

Starting with sundays, they're a bit lazy going, having very active minutes quite low even though sedentary minutes are average.

Mondays are quite wierd, even though sedentary minutes are the its highst value of the week, the amount of activity is still pretty high, that probably comes from the fact that people are well rested from the weekend, have a lot of energy to spare, and are not really busy judging by the amounts of rest and activity they manage to do in the day, probably becouse possible commitments got taken kare of in the weekend.

Tuesday are similar do mandays, and even though the amount of sedentary minutes are starting to go down, clients are still motivated to stay active.

Wednesdays are as average as it gets, both the variables are average.

Thursdays are the opposite of mondays, probably becouse it's the middle of thee week, people are really busy and they're starting to get really tired and unmotivated.

Fridays are definetly the worst one, the sedentary minutes are above the median, while the very active minutes are far below the median, probably becouse since it's the end of the week the clients are the most tired and unmotivated.

This hypotesis is enphasized by saturday, which is where sedentary minutes are almost at their lowest while very active minutes are above the median, customers are motivated again.


We can see a trend with these bar chart: motivation starts really high at the start of the week with lots of activity and gradually goes down until the weekend comes, and even though people don't really want to workout sunday, they still regain energy to start working out the next week.
