# <p id = "b3"><strong><h2>PROCESS</h2></strong></p>
Time to clean and organize the different dataset in order to have a better understanding and easier access to the crucial parts of the data.


#### Cleaning the datasets.

Renaming all the names of the variables of the used datasets to lowercase for better integrity.

```python
daily_activity.columns = map(str.lower, daily_activity.columns)
heartrate_seconds.columns = map(str.lower, heartrate_seconds.columns)
sleep_day.columns = map(str.lower, sleep_day.columns)
```

Cleaning the "daily_activities" datasets from eventual NA values and modify the "activitydate" column to a date format.

```python
daily_activity['activitydate'] = pd.to_datetime(daily_activity['activitydate'], format = '%m/%d/%Y')
daily_activity.rename(columns={'activitydate' : 'date'}, inplace=True)
daily_activity = daily_activity.dropna()
```

Cleaning the "heartrate_seconds" datasets from eventual NA values, modify the "Time" column to a date format and summarize the dataset by calculating the mean of the heart rate throughout the day.

```python
heartrate_seconds[["date", "time", "AM/PM"]] = heartrate_seconds["time"].str.split(" ", expand = True)
heartrate_seconds['date'] = pd.to_datetime(heartrate_seconds['date'], format = '%m/%d/%Y')
heartrate_seconds.groupby(["id", "date"]).first()

concise_heartrate = heartrate_seconds[["id", "date", "value"]]
concise_heartrate = concise_heartrate.set_index('id').groupby('id').resample('D', on = 'date').mean().reset_index()
```

Dividing the "sleep_day" dataset into the ones that slept between the 7-9 hours of recommended sleep, and the ones who didn't, cleaning both datasets from eventual NA values and modify the "sleepday" column to a date format.

```python
sleep_day[["date", "time", "AM/PM"]] = sleep_day["sleepday"].str.split(" ", expand = True)
sleep_day['date'] = pd.to_datetime(sleep_day['date'], format = '%m/%d/%Y')

good_sleep = sleep_day[(sleep_day.totalminutesasleep >= 420) & (sleep_day.totalminutesasleep <= 540)]
good_sleep = good_sleep.drop(['sleepday'], axis=1)

bad_sleep = sleep_day[(sleep_day.totalminutesasleep <= 420) | (sleep_day.totalminutesasleep >= 540)]
bad_sleep = bad_sleep.drop(['sleepday'], axis=1)
```


#### Merging the datasets.
Merging the datasets in order to have more relevant information for the "Leaf" product and marketing analysis.

```python
heartSum_goodsleep = pd.merge(concise_heartrate, good_sleep)
heartSum_goodsleep = heartSum_goodsleep.dropna()

heartSum_badsleep = pd.merge(concise_heartrate, bad_sleep)
heartSum_badsleep = heartSum_badsleep.dropna()

activity_heartrate = pd.merge(daily_activity, concise_heartrate)
activity_heartrate = activity_heartrate.dropna()
```

So I made a bar chart comparing the various days of the week in terms of sedentary minutes and very active minutes.

```python
#quale giorno è il piu sedentario

    #adding day of week to daily_activity dataset 
daily_activity['weekday'] = daily_activity['date'].dt.strftime('%a')


    #standardizing data so variations are confrontable
sed_ver_min = daily_activity[['sedentaryminutes', 'veryactiveminutes', 'veryactivedistance' , 'weekday']]

stand_sed = (sed_ver_min['sedentaryminutes'] - sed_ver_min['sedentaryminutes'].mean()) / sed_ver_min['sedentaryminutes'].std()
stand_sed = pd.concat([stand_sed, sed_ver_min['weekday']], axis = 1)

stand_ver = (sed_ver_min['veryactiveminutes'] - sed_ver_min['veryactiveminutes'].mean()) / sed_ver_min['veryactiveminutes'].std()
stand_ver = pd.concat([stand_ver, sed_ver_min['weekday']], axis = 1)

stand_dist = (sed_ver_min['veryactivedistance'] - sed_ver_min['veryactivedistance'].mean()) / sed_ver_min['veryactivedistance'].std()
stand_dist = pd.concat([stand_dist, sed_ver_min['weekday']], axis = 1)

    #sorting data
a = stand_sed.groupby('weekday').mean()
a = a.loc[['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'], :]

b = stand_ver.groupby('weekday').mean()
b = b.loc[['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'], :]

c = stand_dist.groupby('weekday').mean()
c = c.loc[['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'], :]

y = a['sedentaryminutes']
y1 = b['veryactiveminutes']
y2 = c['veryactivedistance']
x = a.index

X_axis = np.arange(len(x))

    #plotting bar chart
plt.grid()
plt.bar(X_axis - 0.2, y, 0.4, label = 'Sedentary Min.')
plt.bar(X_axis + 0.2, y1, 0.4, label = 'Very Vctive Min.')

plt.xticks(X_axis, x)
plt.xlabel("Day of Week")
plt.ylabel("Deviation From the Mean")
plt.title("Laziest / Most Active Day")
plt.legend()
plt.show()

```



With the bar chart, which was standardized since sedentary minutes and very active minutes have much differents average values, we can clearly see that there are some days worst than others in terms of motivation.

Starting with sundays, they're a bit lazy going, having very active minutes quite low even though sedentary minutes are average.

Mondays are quite wierd, even though sedentary minutes are the its highst value of the week, the amount of activity is still pretty high, that probably comes from the fact that people are well rested from the weekend, have a lot of energy to spare, and are not really busy judging by the amounts of rest and activity they manage to do in the day, probably becouse possible commitments got taken kare of in the weekend.

Tuesday are similar do mandays, and even though the amount of sedentary minutes are starting to go down, clients are still motivated to stay active.

Wednesdays are as average as it gets, both the variables are average.

Thursdays are the opposite of mondays, probably becouse it's the middle of thee week, people are really busy and they're starting to get really tired and unmotivated.

Fridays are definetly the worst one, the sedentary minutes are above the median, while the very active minutes are far below the median, probably becouse since it's the end of the week the clients are the most tired and unmotivated.

This hypotesis is enphasized by saturday, which is where sedentary minutes are almost at their lowest while very active minutes are above the median, customers are motivated again.


We can see a trend with these bar chart: motivation starts really high at the start of the week with lots of activity and gradually goes down until the weekend comes, and even though people don't really want to workout sunday, they still regain energy to start working out the next week.
